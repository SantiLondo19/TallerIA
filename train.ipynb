```python
import os
```

```python
base_dir = r'C:\Users\lola1\Downloads\PetImages\base_dir'
train_dir = os.path.join(base_dir, 'train_dir')
test_dir = os.path.join(base_dir, 'test_dir')
validation_dir = os.path.join(base_dir, 'validation_dir')
```

```python
train_cats_dir = os.path.join(train_dir, 'train_cats_dir')
train_dogs_dir = os.path.join(train_dir, 'train_dogs_dir')
validation_cats_dir = os.path.join(validation_dir, 'validation_cats_dir')
validation_dogs_dir = os.path.join(validation_dir, 'validation_dogs_dir')
test_dogs_dir = os.path.join(test_dir, 'test_dogs_dir')
test_cats_dir = os.path.join(test_dir, 'test_cats_dir')
```

```python
print('Total gatos entrenamiento:', len(os.listdir(train_cats_dir)))
print('Total perros entrenamiento:', len(os.listdir(train_dogs_dir)))
print('Total perros validacion:', len(os.listdir(validation_dogs_dir)))
print('Total gatos validacion:', len(os.listdir(validation_cats_dir)))
print('Total perros test:', len(os.listdir(test_dogs_dir)))
print('Total gatos test:', len(os.listdir(test_cats_dir)))
```

```python
from keras.preprocessing.image import ImageDataGenerator
#  es una herramienta muy útil para realizar aumentación de datos y preprocesamiento de imágenes en el contexto de entrenamiento de modelos de aprendizaje profundo. 
#Esta línea importa la clase ImageDataGenerator del módulo keras.preprocessing.image. Esta clase proporciona métodos para generar lotes de datos de imágenes con transformaciones y aumentación aplicadas a cada imagen.
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='binary')
'''
Aquí se crean dos instancias de ImageDataGenerator, train_datagen y test_datagen. La instrucción rescale=1./255 se utiliza para escalar los valores de píxeles de las imágenes de 0 a 255 a un rango de 0 a 1. Esta es una operación de preprocesamiento común que ayuda a normalizar los datos de entrada para que sean más adecuados para el entrenamiento de modelos de aprendizaje profundo.
Estas dos instrucciones generan generadores de datos que se utilizarán durante el entrenamiento y la validación del modelo. Estos generadores de datos cargarán las imágenes directamente desde los directorios train_dir y validation_dir. Cada generador generará lotes de imágenes con las transformaciones especificadas.

En el caso de train_generator, se especifica el directorio de entrenamiento train_dir, el tamaño objetivo de las imágenes (150, 150), el tamaño del lote batch_size=20 y class_mode='binary' para indicar que se trata de un problema de clasificación binaria. Este generador generará lotes de imágenes de entrenamiento para su uso durante el entrenamiento del modelo.

En el caso de validation_generator, se especifica el directorio de validación validation_dir, el tamaño objetivo de las imágenes (150, 150), el tamaño del lote batch_size=20 y también class_mode='binary'. Este generador generará lotes de imágenes de validación para su uso durante la evaluación del modelo.
'''

```

```python
for data_batch, labels_batch in train_generator:
    print( 'Forma del tensor de datos', data_batch.shape)
    print ('Forma del tensor de etiquetas', labels_batch.shape)
    print (data_batch)
    print(labels_batch)
    break
    
'''

Este código implementa un bucle for que itera sobre los lotes de datos generados por el objeto train_generator.

En cada iteración del bucle, se cargan batch_size imágenes y etiquetas del conjunto de datos de entrenamiento y se almacenan en las variables data_batch y labels_batch, respectivamente. Luego, se imprime la forma de los tensores de datos y etiquetas utilizando las funciones shape.

La función shape devuelve una tupla que indica las dimensiones del tensor. En este caso, la forma del tensor de datos es (batch_size, 150, 150, 3), lo que significa que hay batch_size imágenes de tamaño 150x150 píxeles con 3 canales de color (RGB). La forma del tensor de etiquetas es (batch_size,), lo que significa que hay batch_size etiquetas binarias para las imágenes cargadas.

A continuación, se imprimen los datos y etiquetas de los lotes utilizando las variables data_batch y labels_batch. Finalmente, se utiliza la instrucción break para salir del bucle después de procesar el primer lote.
'''
```

```python
import matplotlib.pyplot as plt

# Seleccionar una imagen aleatoria del tensor data_batch
img = data_batch[1]

# Mostrar la imagen utilizando la función imshow de matplotlib
plt.imshow(img)
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definición del modelo
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Resumen del modelo
model.summary()

# Compilación del modelo
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_generator,steps_per_epoch = 100, epochs = 30, validation_data = validation_generator, validation_steps = 50)
```

```python
# Imprimir la precisión (accuracy) del conjunto de entrenamiento y validación
print("Última precisión del conjunto de entrenamiento:", history.history['accuracy'][-1])
print("Última precisión del conjunto de validación:", history.history['val_accuracy'][-1])

# Imprimir la pérdida (loss) del conjunto de entrenamiento y validación
print("Última pérdida del conjunto de entrenamiento:", history.history['loss'][-1])
print("Última pérdida del conjunto de validación:", history.history['val_loss'][-1])
```

```python
import matplotlib.pyplot as plt

# Obtener las métricas de entrenamiento y validación
train_loss = history.history['loss']
train_accuracy = history.history['accuracy']
val_loss = history.history['val_loss']
val_accuracy = history.history['val_accuracy']

# Generar gráfica de pérdida
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Generar gráfica de precisión
plt.plot(range(1, len(train_accuracy) + 1), train_accuracy, label='Training Accuracy')
plt.plot(range(1, len(val_accuracy) + 1), val_accuracy, label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
